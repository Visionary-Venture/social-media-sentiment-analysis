## üìå Project Overview
This report analyzes customer sentiment across major airlines using Twitter data. The goal is to uncover trends, identify pain points, and provide actionable recommendations to improve engagement and brand perception.


```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

df = pd.read_csv("../data/processed/tweets_clean.csv")
df['clean_text'] = df['clean_text'].fillna("").astype(str)
```

```python
df['airline_sentiment'].value_counts().plot(kind='bar', color=['tomato','gold','steelblue'])
plt.title('Sentiment Distribution (Cleaned)')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig("../docs/sentiment_distribution.png")
plt.show()
```

### üìä Sentiment Distribution
Most tweets express negative sentiment, indicating widespread dissatisfaction. Positive feedback is limited, suggesting room for improvement in customer experience.


```python
import seaborn as sns

sns.countplot(x="airline", hue="airline_sentiment", data=df,
              palette={"negative":"tomato","neutral":"gold","positive":"steelblue"})
plt.title("Sentiment by Airline")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../docs/sentiment_by_airline.png")
plt.show()
```

### ‚úàÔ∏è Sentiment by Airline
United, US Airways, and American Airlines received the most negative tweets. Southwest and Delta show more balanced sentiment, while Virgin America has the fewest mentions.


```python
df['tweet_created'] = pd.to_datetime(df['tweet_created'])
df.groupby(df['tweet_created'].dt.date)['airline_sentiment'].value_counts().unstack().plot(kind='line')
plt.title("Sentiment Trend Over Time")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../docs/sentiment_trend.png")
plt.show()
```

### üìà Sentiment Trend Over Time
A spike in negative sentiment occurred around February 22, likely due to delays or cancellations. Monitoring such spikes can help airlines respond proactively.


```python
text = " ".join(df['clean_text'])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig("../docs/wordcloud.png")
plt.show()
```

### ‚òÅÔ∏è Word Cloud of Tweets
Common concerns include flight delays, cancellations, customer service, and long wait times. Positive mentions include gratitude and helpful staff.
